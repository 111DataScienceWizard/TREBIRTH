```python
import pandas as pd
import numpy as np
from scipy.stats import skew
from scipy import signal
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
def detrend_raw_columns(df):
    detrended_columns = []
    for raw_column in df.columns:
        df[raw_column] = df[raw_column] - df[raw_column].mean()
        detrended_columns.append(raw_column)
    return df[detrended_columns]
def calculate_statistics(dataframe):
  for col in dataframe:
    stats_df = pd.DataFrame({
        'Std': dataframe.std(),
        'PTP': np.ptp(dataframe),
    })
    return stats_df
def fq(df):
    frequencies = []
    powers = []

    for i in df:
        f, p = signal.welch(df[i], 100, 'flattop', 1024, scaling='spectrum')
        frequencies.append(f)
        powers.append(p)

    frequencies = pd.DataFrame(frequencies)
    powers = pd.DataFrame(powers)
    return frequencies,powers

def calculate_rolling_statistics(dataframe, window_size):
    rolling_stats_dfs = []

    for col in dataframe:
        col_std = dataframe[col].rolling(window=window_size, min_periods=1).std()
        col_ptp = dataframe[col].rolling(window=window_size, min_periods=1).apply(np.ptp)
        col_stats_df = pd.DataFrame({
            'Std': col_std,
            'PTP': col_ptp
        })
        rolling_stats_dfs.append(col_stats_df)
    concatenated_df = pd.concat(rolling_stats_dfs, axis=1)
    return concatenated_df.dropna()
def melt_and_pivot(df):
    df_melted = pd.melt(df, value_vars=['Std', 'PTP'], var_name='Variable', value_name='Value')
    df_melted['Type'] = df_melted.groupby('Variable').cumcount()
    df_result = df_melted.pivot_table(index=['Type'], columns=['Variable'], values='Value', aggfunc='first').reset_index(drop=True)
    df_result.columns = ['Std', 'PTP']
    return df_result

def process_excel_file(file_path):
    data = pd.read_excel(file_path, header=None)
    data_df = pd.DataFrame(data)
    data_detrend = detrend_raw_columns(data)
    data_f, data_p = fq(data)
    data_calculate_rolling_statistics = calculate_rolling_statistics(data, 100)
    data_melt_and_pivot = melt_and_pivot(data_calculate_rolling_statistics).dropna()
    data_p_T = pd.DataFrame(data_p.T).dropna()
    data_p = pd.DataFrame(data_p)
    data_p_calculate_rolling_statistics = calculate_rolling_statistics(data_p, 100)
    data_p_melt_and_pivot = melt_and_pivot(data_p_calculate_rolling_statistics).dropna()
    data_p_calculate_statistics = calculate_statistics(data_p).dropna()
    data_detrend_f, data_detrend_p = fq(data_detrend)
    data_detrend_calculate_rolling_statistics = calculate_rolling_statistics(data_detrend, 100)
    data_detrend_melt_and_pivot = melt_and_pivot(data_detrend_calculate_rolling_statistics).dropna()
    data_detrend_p = pd.DataFrame(data_detrend_p)
    data_detrend_p_T = pd.DataFrame(data_detrend_p.T).dropna()
    data_detrend_p_calculate_rolling_statistics = calculate_rolling_statistics(data_detrend_p, 100)
    data_detrend_p_melt_and_pivot = melt_and_pivot(data_detrend_p_calculate_rolling_statistics).dropna()
    data_detrend_p_calculate_statistics = calculate_statistics(data_detrend_p).dropna()
    return (        data_df.dropna(),
        data_melt_and_pivot,
        data_p_T,
        data_p,
        data_p_melt_and_pivot,
        data_p_calculate_statistics,
        data_detrend_melt_and_pivot,
        data_detrend_p_T,
        data_detrend_p,
        data_detrend_p_melt_and_pivot,
        data_detrend_p_calculate_statistics
    )
```
